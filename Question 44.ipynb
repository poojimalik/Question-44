```python
def swap_numbers(a,b):
    a,b = b,a
    return a,b
if __name__ =="__main__":
    try:
        a = 5
        b = 10
        print("a before swap :",a)
        print("b before swap :",b)
        a,b = swap_numbers(a,b)
        print("a after swap :",a)
        print("b after swap :",b)
     except Exception as e:
         print(e)
```
